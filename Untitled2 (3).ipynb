#### **Outlier detection and treatment:**






```python
# Create a boxplot for the continuous variables to observe outliers

LaptopData.boxplot(column=['Processor_Speed', 'Screen_Size', 'Weight'], figsize=(18,10))
```

#### **Observations from Boxplots**
**Processor_Speed:**
* There are a few outliers with processor speeds above 4.0Ghz.

**Screen_Size:**

**Weight:**

```python
# Removal of Outliers from dataset
LaptopData_Clean = LaptopData.copy()

LaptopData_Clean = LaptopData_Clean[LaptopData_Clean['Processor_Speed'] < 4.0]
LaptopData_Clean = LaptopData_Clean[LaptopData_Clean['Screen_Size'] < 17.0]
LaptopData_Clean = LaptopData_Clean[LaptopData_Clean['Weight'] < 4.0]
```

### **Step 12: Training/Testing Sampling and Standardization**

```python
from sklearn.model_selection import train_test_split

# Separate Target Variables and Predictor Variables
X = DataForML_Numeric.drop('Price', axis=1)
y = DataForML_Numeric['Price']

# Split the data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
### Standardization of data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train_scaled, y_train, test_size=0.2, random_state=42)
```

```python
print('Shape of X_train:', X_train.shape)
print('Shape of X_test:', X_test.shape)
print('Shape of y_train:', y_train.shape)
print('Shape of y_test:', y_test.shape)
```

### **Step 13: Investigating multiple Regression Models**
#### 1. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Create a Random Forest Regressor model
model_rf = RandomForestRegressor()

# Train the model
model_rf.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = model_rf.predict(X_test)

# Evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
print("Mean squared error:", mse_rf)

# Print the model summary
print(model_rf)
```

#### 2. Linear Regression

```python
from sklearn.linear_model import LinearRegression
# Create a Linear Regression model
model_lr = LinearRegression()

# Train the model
model_lr.fit(X_train, y_train)

# Make predictions on the test set
y_pred_lr = model_lr.predict(X_test)

# Evaluate the model
mse_lr = mean_squared_error(y_test, y_pred_lr)
print("Mean squared error:", mse_lr)

# Print the model summary
print(model_lr)
```

#### 3. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

# Create a Decision Tree Regressor model
model_dt = DecisionTreeRegressor()

# Train the model
model_dt.fit(X_train, y_train)

# Make predictions on the test set
y_pred_dt = model_dt.predict(X_test)

# Evaluate the model
mse_dt = mean_squared_error(y_test, y_pred_dt)
print("Mean squared error:", mse_dt)

# Print the model summary
print(model_dt)
```

### **Step 14: Selection of the best model**

* The selected model is Linear Regression as it returns the lowest Mean squared error, therefore making it the most accurate.

```python
TargetVariable = 'Price'

Predictors = ['Storage_Capacity', 'RAM_Size', 'Weight', 'Screen_Size', 'Processor_Speed']

X = DataForML_Numeric[Predictors].values
y = DataForML_Numeric[TargetVariable].values

### Standardization of data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scalerFit = scaler.fit(X)

X = scalerFit.transform(X)

print(X.shape)
print(y.shape)
```

### **Step 15: Deployment of the Model**
**Command Based Program to test the function:**

```python

# Import necessary libraries
import pickle
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the trained model and scaler
model = pickle.load(open('FinalModel.pkl', 'rb'))
scaler = pickle.load(open('scaler.pkl', 'rb'))

def predict_price(storage_capacity, ram_size, weight, screen_size, processor_speed):

  # Create a DataFrame with the input data
  input_data = pd.DataFrame([[storage_capacity, ram_size, weight, screen_size, processor_speed]], columns=['Storage_Capacity', 'RAM_Size', 'Weight', 'Screen_Size', 'Processor_Speed'])

  # Get the dummies for the categorical variables
  input_data = pd.get_dummies(input_data)

  # Scale the data
  input_data_scaled = scaler.transform(input_data)

  # Predict the price
  predicted_price = model.predict(input_data_scaled)[0]

  # Return the predicted price
  return predicted_price

# Get user input
storage_capacity = int(input("Enter the storage capacity of the laptop (in GB): "))
ram_size = int(input("Enter the RAM size of the laptop (in GB): "))
weight = float(input("Enter the weight of the laptop (in kg): "))
screen_size = float(input("Enter the screen size of the laptop (in inches): "))
processor_speed = float(input("Enter the processor speed of the laptop (in GHz): "))

# Predict the price
predicted_price = predict_price(storage_capacity, ram_size, weight, screen_size, processor_speed)

# Print the predicted price
print("The predicted price of the laptop is: $", predicted_price)
```

**GUI Program Deployment:**

Please note: The GUI program will not run on Google Colab, must be exported into Visual Studio or Pycharm. The CSV file needs to be opened locally as it will not connect to google drive.

```python
import tkinter
from tkinter import messagebox


model = pickle.load(open('FinalModel.pkl', 'rb'))
scaler = pickle.load(open('scaler.pkl', 'rb'))

class LaptopPricePredictionApp:
    def __init__(self):
        self.main_window = tkinter.Tk()
        self.main_window.title("Laptop Price Prediction")

        # Entry frames and price result frame
        self.frame0 = tkinter.Frame()
        self.frame1 = tkinter.Frame()
        self.frame2 = tkinter.Frame()
        self.frame3 = tkinter.Frame()
        self.frame4 = tkinter.Frame()
        self.btnframe = tkinter.Frame()
        self.frame5 = tkinter.Frame()

        # Top frame widgets
        storage_lbl = tkinter.Label(self.frame0, text="Storage Capacity (in GB)")
        self.storage_entry = tkinter.Entry(self.frame0, bg='light grey', bd=2, width='10')
        storage_lbl.pack(side='left')
        self.storage_entry.pack(side='left')

        # Frame 1 widgets
        ram_lbl = tkinter.Label(self.frame1, text="Ram Size (in GB)" )
        self.ram_entry = tkinter.Entry(self.frame1, bg='light grey', bd=2, width='10')
        ram_lbl.pack(side='left')
        self.ram_entry.pack(side='left')

        # Frame 2 widgets
        weight_lbl = tkinter.Label(self.frame2, text="Weight (in KG)" )
        self.weight_entry = tkinter.Entry(self.frame2, bg='light grey', bd=2, width='10')
        weight_lbl.pack(side='left')
        self.weight_entry.pack(side='left')

        # Frame 3 widgets
        screen_size_lbl = tkinter.Label(self.frame3, text="Screen Size (in inches)" )
        self.screen_size_entry = tkinter.Entry(self.frame3, bg='light grey', bd=2, width='10')
        screen_size_lbl.pack(side='left')
        self.screen_size_entry.pack(side='left')

          # Frame 4 widgets
        processor_speed_lbl = tkinter.Label(self.frame4, text="Processor Speed (in GHZ)" )
        self.processor_speed_entry = tkinter.Entry(self.frame4, bg='light grey', bd=2, width='10')
        processor_speed_lbl.pack(side='left')
        self.processor_speed_entry.pack(side='left')

        # Btn frame widgets
        predict_price_btn = tkinter.Button(self.btnframe, text="Predict Price", command=self.predict_price)
        quit_btn = tkinter.Button(self.btnframe, text="Quit", command=self.main_window.destroy)
        predict_price_btn.pack(side='left')
        quit_btn.pack(side='left')

          # Result widgets
        self.predicted_price_result = tkinter.Text(self.frame5, bg='light blue')
        self.predicted_price_result.pack()

        # Pack the frames
        self.frame0.pack()
        self.frame1.pack()
        self.frame2.pack()
        self.frame3.pack()
        self.frame4.pack()
        self.btnframe.pack()
        self.frame5.pack()

        tkinter.mainloop()

    def predict_price(self):
        result_string = ""
        self.predicted_price_result.delete('1.0', tkinter.END)

        storage_capacity = int(self.storage_entry.get())
        ram_size = int(self.ram_entry.get())
        weight = float(self.weight_entry.get())
        screen_size = float(self.screen_size_entry.get())
        processor_speed = float(self.processor_speed_entry.get())

        input_data = pd.DataFrame([storage_capacity, ram_size, weight, screen_size, processor_speed], columns=['Storage_Capacity', 'RAM_Size', 'Weight', 'Screen_Size', 'Processor_Speed'])

        input_data = pd.get_dummies(input_data)

        input_data_scaled = scaler.transform(input_data)

        predicted_price = model.predict(input_data_scaled)

        result_string+= predicted_price

        messagebox.info("Predicted Price", f"The predicted price of the laptop is: ${result_string}")







app = LaptopPricePredictionApp()
```
